```python
date_wanted = 'all' #YYYY-MM-DD or all
job_page = 'openings.htm'
```

```python
from bs4 import BeautifulSoup
import re
import json
import pandas as pd
```

```python
nearby_high_schools = open('nearby_high_schools.txt').read().split('\n')
nearby_k_8 = open('nearby_k_8.txt').read().split('\n')
train_bus_high_schools = open('train_bus_high_schools.txt').read().split('\n')
train_bus_k_8 = open('train_bus_k_8.txt').read().split('\n')
```

```python
all_nearby = nearby_high_schools + nearby_k_8
all_train_bus = train_bus_high_schools + train_bus_k_8
all_schools = nearby_high_schools + nearby_k_8 + train_bus_high_schools + train_bus_k_8
```

```python
zips = pd.read_csv('Neighboorhood_zips.csv',header=None,names=['Zip','Neighborhoods'])
```

```python
soup = BeautifulSoup(open(job_page),'html.parser')
for script in soup.find_all('script'):
    if 'availJobs' in str(script):
        m = re.search('availJobs: ({.*})',str(script))
        jobs = json.loads(m.group(1))
```

```python
jobs['list'][8]
```

```python
jobs_df = pd.DataFrame(columns=['Nearby','Train/Bus','School','Date','Start_time','End_time','Multiday','Has_notes',
                                'Teacher_first_name','Teacher_last_name','Address','Zip'])

for job in jobs['list']:
    multiday = False
    if len(job['Items']) != 1:
        multiday = True
    school = job['Items'][0]['Institution']['Name']
    if school in all_nearby:
        nearby = 'Y'
        train_bus = 'N'
    elif school in all_train_bus:
        nearby = 'N'
        train_bus = 'Y'
    else:
        nearby = 'N'
        train_bus = 'N'
    date = job['Items'][0]['Start'].split('T')[0]
    start_time = job['Items'][0]['Start'].split('T')[1]
    end_time = job['Items'][0]['End'].split('T')[1]
    first_name = job['WorkerFirstName']
    last_name = job['WorkerLastName']
    notes = job['HasNotes']
    address = job['Items'][0]['Institution']['Street1']
    zip_code = int(job['Items'][0]['Institution']['Zip'].strip())
    
    if job['WorkerTitle'] == 'Regular Teacher' or job['WorkerTitle'] == 'Program Option Teacher':
        jobs_df = jobs_df.append(pd.DataFrame([[nearby, train_bus, school, date, start_time, end_time, multiday,
                                            notes, first_name, last_name, address, zip_code]], 
                                          columns=['Nearby', 'Train/Bus', 'School', 'Date', 'Start_time', 'End_time', 'Multiday',
                                                   'Has_notes', 'Teacher_first_name', 'Teacher_last_name', 'Address', 'Zip']))
    
    
    
jobs_df = jobs_df.merge(zips, on='Zip').sort_values(['Nearby', 'Train/Bus'],ascending=False).reset_index(drop=True)
if date_wanted == 'all':
    tomorrow = jobs_df
else:
    tomorrow = jobs_df[jobs_df['Date'] == date_wanted]
tomorrow
```
